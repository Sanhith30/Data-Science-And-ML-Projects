```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('/content/Hyderabad_House_Data.csv')
```

```python
actual_data = data.copy()
display(actual_data.head(10))
```

```python
studio_apartment_rows = actual_data[actual_data['Bedrooms'].str.contains('Studio Apartment', na=False)]
apartment_rows = actual_data[actual_data['Bedrooms'].str.contains('Apartment', na=False)]

print("Rows with 'Studio Apartment' in Bedrooms:")
display(studio_apartment_rows)

print("\nRows with 'Apartment' in Bedrooms:")
display(apartment_rows)
```

```python
data = data.drop('Unnamed: 0', axis=1)
```

```python
data
```

```python
for col in data.columns:
  display(data[col].value_counts())
```

```python
data['Bedrooms'] = data['Bedrooms'].replace('Studio Apartment', '1', regex=True)
data['Bedrooms'] = data['Bedrooms'].str.replace(r'[^\d]', '', regex=True)
data['Bedrooms'] = pd.to_numeric(data['Bedrooms'], errors='coerce')
data.dropna(subset=['Bedrooms'], inplace=True)
display(data['Bedrooms'].value_counts())
```

```python

non_numeric_bathroom_rows = actual_data[pd.to_numeric(actual_data['Bathrooms'], errors='coerce').isna()]
display(non_numeric_bathroom_rows)
```

```python
data['Bedrooms'].isnull().sum()
```

```python
data['Bathrooms'] = data['Bathrooms'].astype(str).str.replace(r'[^\d]', '', regex=True)
data['Bathrooms'] = pd.to_numeric(data['Bathrooms'], errors='coerce')
display(data['Bathrooms'].value_counts())
```

```python
data=data[data['Bathrooms'] != 19]
```

```python
display(data['Bathrooms'].isnull().sum())
display(data['Bathrooms'].value_counts())
```

```python
data['Bathrooms'] = data.apply(lambda row: row['Bedrooms'] if pd.isnull(row['Bathrooms']) else row['Bathrooms'], axis=1)
```

```python
display(data['Bathrooms'].isnull().sum())
```

```python
data
```

```python
desired_furnishing_values = ['Unfurnished', 'Semi-Furnished', 'Furnished']
data = data[data['Furnishing'].isin(desired_furnishing_values)]
display(data['Furnishing'].value_counts())
```

```python
data
```

```python
undesired_tennants_values = ['2', '3', '1']
data = data[~data['Tennants'].isin(undesired_tennants_values)]

display(data['Tennants'].value_counts())
```

```python
display(data['Tennants'].isnull().sum())
```

```python
data = data.rename(columns={'Area': 'Area in sqft'})
```

```python
data['Area in sqft'] = data['Area in sqft'].astype(str).str.replace(r'[^\d]', '', regex=True)
data['Area in sqft'] = pd.to_numeric(data['Area in sqft'], errors='coerce')
display(data['Area in sqft'].isnull().sum())
```

```python
data['Area in sqft'] = data.groupby('Bedrooms')['Area in sqft'].transform(lambda x: x.fillna(x.mean()))
display(data['Area in sqft'].isnull().sum())
```

```python
data['Price'] = data['Price'].astype(str).str.replace(r'[^\d]', '', regex=True)
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
display(data['Price'].isnull().sum())
```
